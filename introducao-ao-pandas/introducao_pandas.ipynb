**Introdução ao Pandas no Python**

Importando Pandas

```python
import pandas as pd
```

**Criando um *dataframe* a partir de um dicionário**
- dataframe é uma tabela dentro do Python

```python
dataframe = pd.DataFrame() # cria uma tabela vazia

# criando um dicionário
venda = {'data': ['15/02/2021', '16/02/2021'],
       'valor': [500, 300],
       'produto': ['feijao', 'arroz'],
       'qtde': [50, 70]
}

# criando uma dataframe a partir de um dicionario
vendas_df = pd.DataFrame(venda)
```

**Visualização dos dados**
- *print*: exibe como se fosse um bloco de texto
- *display*: exibe uma tabela mais estilizada (melhor visualização)

```python
print(vendas_df)
display(vendas_df)
```

**Importando arquivos e base de dados**

```python
# transforma o arquivo em dataframe
vendas_df = pd.read_excel("Vendas.xlsx")
display(vendas_df)
```

**Visualização de dados**
- head: mostra as quantidades de linhas informada
- shape: mostra quantas linhas e colunas tem a tabela (menos custoso que o display)
- describe: faz um resumo das colunas numéricas

```python
display(vendas_df.head(5))
print(vendas_df.shape)
display(vendas_df.describe())
```

**Visualizar 1 coluna**

```python
produtos = vendas_df[['Produto', 'ID Loja']]
display(produtos)
```

**Método .loc**
- Pega uma linha específica

```python
# pegar uma linha específica
display(vendas_df.loc[1:5])
```

- Faz seleção nos dataframes

```python
# pegar linhas e colunas do loc
# .loc[linhas, colunas]
vendas_norteshopping_df = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping', ['ID Loja', 'Produto', 'Quantidade']]
```

- Pega linhas de acordo com alguma condição

```python
# pegar linha que corresponde a uma condição
vendas_norteshopping_df = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping']
```


- Pegar um valor específico

```python
# pegar um valor específico
print(vendas_df.loc[1, 'Produto'])
```

**Adicionando uma coluna**
- A partir de uma coluna existente
- Criar uma nova com valor padrão (menos custoso)

```python
vendas_dez_df = pd.read_excel("Vendas - Dez.xlsx")

vendas_df = pd.concat([vendas_df, vendas_dez_df])
display(vendas_df)
```

```python
# a partir de uma coluna existente
vendas_df['Comissão'] = vendas_df['Valor Final'] * 0.05
# display(vendas_df)

# criar uma coluna com valor padrão
vendas_df.loc[:, 'Imposto'] = 0
display(vendas_df)
```

```python
vendas_df = vendas_df.drop(0, axis=0)
display(vendas_df)
```

**Comandos importantes para análise de dados**  
*Valores vazios*
- Detectar linhas/colunas vazias
- Deletar linhas que possuem valores vazios
- Preencher valores vazios (média e último valor)

```python
# deletar linhas e colunas totalmente vazias
vendas_df = vendas_df.dropna(how='all', axis=1)
display(vendas_df)

# deletar linhas que possuem pelo menos um valor vazio
# vendas_df = vendas_df.dropna()

# prencher valores vazios
# média
vendas_df['Comissão'] = vendas_df['Comissão'].fillna(vendas_df['Comissão'].mean()) 

# último valor
# vendas_df = vendas_df.ffill()
```

**Calcular indicadores**
- Groupby
- Value counts

```python
# value counts
transacoes_lojas = vendas_df['ID Loja'].value_counts()
display(transacoes_lojas)

#groupby
faturamento_produto = vendas_df[['Produto', 'Valor Final']].groupby('Produto').sum()
display(faturamento_produto)
```

```python
faturamento_loja = vendas_df[['ID Loja', 'Valor Final']].groupby('ID Loja').sum()
display(faturamento_loja)
```

**Mesclar dois dataframes**  
Procurar informações de um dataframe em outro

```python
gerentes_df = pd.read_excel("Gerentes.xlsx")

# objetivo: mostrar no dataframe vendas_df os gerentes
vendas_df = vendas_df.merge(gerentes_df)
display(vendas_df)
```
